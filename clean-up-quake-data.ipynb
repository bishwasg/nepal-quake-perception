```python
import numpy as np
import pandas as pd
# read the excel data file
data = pd.read_csv("earthquake_data.csv", index_col = 'id')
data.info()
```

```python
data['age'].unique()
```

```python
selected = ['are_main_problems_addressed',
            'satisfied_with_gov', 
            'satisfied_with_ngos',
            'is_support_fair',
            'information_about_gov_support',
            'prepared_for_winter', 
            'have_been_heard',
            'relief_effort_progress',
           ]
mapper = {'1___not_at_all':1, 
          '2___very_little':2, 
          '2___not_very_m':2, 
          '3___neutral':3, 
          '4___mostly_yes':4,  
          '5___completely_yes':5,
          '5___completely':5,
          'refused': np.nan,
          'don_t_know': np.nan}
for cat in selected:
  data[cat].replace(mapper,inplace=True) 
    
data[selected]
```
